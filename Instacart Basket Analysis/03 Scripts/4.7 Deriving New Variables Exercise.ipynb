## Define functions

```python
# Create IF function for prices row, filtering the 3 catagoies

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the function

df['price_range'] = df.apply(price_label, axis=1)
```

# Create new busiest_period_of_day column

```python
# Check the order_time column for metrics

df_ords_prods_merge['order_time'].describe()
```

```python
print(df_ords_prods_merge['order_time'].value_counts().sort_index())
```

```python
# Seperate into 3 catagories based on count rank

hour_counts = df_ords_prods_merge['order_time'].value_counts().sort_index()

min_count = hour_counts.min()  # Lowest count
max_count = hour_counts.max()  # Highest count
interval = (max_count - min_count) / 3  # Divide range into 3 parts

lower_threshold = min_count + interval
upper_threshold = max_count - interval
```

```python
# Create new result 

result_new = []
for value in df_ords_prods_merge['order_time']:
    if hour_counts[value] >= upper_threshold:
        result_new.append('Most orders')
    elif hour_counts[value] >= lower_threshold:
        result_new.append('Average orders')
    else:
        result_new.append('Fewest orders')
```

```python
result_new
```

```python
# Assign new busiest period of day column

df_ords_prods_merge['busiest_period_of_day'] = result_new
```

```python
# Print frequency

df_ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```
