```python
# Import Libraries

import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
df_customer = pd.read_csv(os.path.join(path, '02 Data', 'Original Data', 'customers.csv'), index_col = False)
```

## Wrangle Customer Dataframe

```python
# Check columns

df_customer.head()
```

```python
# Check about date_joined consistancy

df_customer['date_joined']
```

```python
# Combined first and last names to one name column with the concat method
# First change values to string, as there are some numerical

df_customer['First Name'] = df_customer['First Name'].astype(str)
df_customer['Surnam'] = df_customer['Surnam'].astype(str)
```

```python
# Combine

df_customer['Full Name'] = df_customer[['First Name', 'Surnam']].agg(' '.join, axis=1)
```

```python
df_customer.head()
```

```python
# Drop Frist Name and Surnam columns

df_customer = df_customer.drop(['First Name', 'Surnam'], axis=1)
```

```python
df_customer.head()
```

```python
# Rename columns

df_customer = df_customer.rename(columns={
    'State': 'state',
    'n_dependants': 'number_of_children',
    'Age': 'age',
    'Gender': 'gender',
    'Full Name': 'name'
})
```

```python
df_customer.head()
```

```python
# Changing variable data type, user_id to string


df_customer['user_id'] = df_customer['user_id'].astype('str')
```

```python
df_customer['user_id'].dtype
```

```python
# Check info for missing values

df_customer.info()
```

```python
df_customer.isnull().sum()
```

## Import prepared Instacart data to merge with customer df

```python
# Import prepared Instacard data to merge with customer df

df_merged = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_prods_merge_updated_grouped.pkl'))
```

```python
df_merged.head()
```

```python
# The user_id column is not matching to the df_customer user_id column as an object/str. Will change it

df_merged['user_id'] = df_merged['user_id'].astype(str)
```

```python
# Drop the already _merged column in df_merged so we can create another one

df_merged = df_merged.drop('_merge', axis=1)
```

## Merge the dataframes

```python
# Merge the two dataframes on the user_id column

df_merged = df_merged.merge(df_customer, on='user_id', how='outer', indicator=True)
```
