## Creating Histograms and Scatterplot

```python
# Create a histogram 

ords_prods_merge['prices'].plot.hist(bins = 25)
```

```python
# Create a scatter plot

sns.scatterplot(x = 'prices', y = 'prices',data = ords_prods_merge)
```

```python
# Checking the reason for the outlier prices


ords_prods_merge[['order_id', 'user_id', 'prices']].loc[ords_prods_merge['prices'] > 100]
```

```python
# Replace prices about 100 to missing values, NaN

ords_prods_merge.loc[ords_prods_merge['prices'] >100, 'prices'] = np.nan
```

```python
ords_prods_merge['prices'].max()
```

```python
# Once cleaned data for histogram

hist = ords_prods_merge['prices'].plot.hist(bins = 70)
```

```python
# Export hist to an image

hist.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'hist_orders_dow.png'))
```

## Create a line chart showing the relationship between age and number of kids

```python
# Use the already created samples for a new object (age, number_of_children)

df_4 = small[['age','number_of_children']]
```
